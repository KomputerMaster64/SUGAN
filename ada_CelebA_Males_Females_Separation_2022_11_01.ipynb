```python
%cd /scratch/temp_44/data/img_align_celeba_test
!ls
%cd /scratch/temp_44/data/img_align_celeba_test/female
!ls
```

```python
import shutil
# male = []
# female = []
mdest_dir = DATA_DIR + '/test/male/'
fdest_dir = DATA_DIR + '/test/female/'

for l in Lines[2:]:
    val = l.split()[21]
    name = l.split()[0]
    src = DATA_DIR + '/img_align_celeba/' + name
    try:
        print(val)
        if val == '-1':
            shutil.copyfile(src, fdest_dir + name)    
        elif val == '1':
            shutil.copyfile(src, mdest_dir + name)     
    except:
        continue
    
```
